```python
sns.set_style("whitegrid")
f, ax = plt.subplots(figsize=[6,3],dpi=300)

plt.grid('both', color='whitesmoke')
ax.plot(sf.Ano,sf['3.2. Agriculture']/100, color="skyblue", marker='.')
# ax.plot(sf.Ano,sf['3.2.1.1. Soybean']/100, color="darkorange", marker='.')

ax.bar(sf.Ano,sf['3.2.1.2. Sugar Cane']/100, 0.8, color="darkorange", edgecolor='none' )
ax.bar(sf.Ano,sf['3.2.1.4. Other Temporary Crops']/100, 0.8, color="red",edgecolor='none', bottom=sf['3.2.1.2. Sugar Cane']/100)
ax.bar(sf.Ano,sf['3.2.2.2. Citrus']/100, 0.8, color="pink",edgecolor='none',bottom=(sf['3.2.1.2. Sugar Cane']+sf['3.2.1.4. Other Temporary Crops'])/100)
ax.bar(sf.Ano,sf['3.2.2.1. Coffee']/100, 0.8, color="mediumorchid",edgecolor='none',bottom=(sf['3.2.1.2. Sugar Cane']+sf['3.2.1.4. Other Temporary Crops']+sf['3.2.2.2. Citrus'])/100)
ax.bar(sf.Ano,sf['3.2.1.1. Soybean']/100, 0.8, color="slateblue",edgecolor='none',bottom=(sf['3.2.1.2. Sugar Cane']+sf['3.2.1.4. Other Temporary Crops']+sf['3.2.2.2. Citrus']+sf['3.2.2.1. Coffee'])/100)

plt.title('')
plt.ylabel('Area ($km^{2}$)')
plt.legend(['Total agriculture','Sugar cane','Other temporary crops','Citrus','Coffee','Soy'],bbox_to_anchor=(1.04,1), loc="upper left")
#plt.show()
#plt.xticks(np.arange(min(sf.Ano), max(sf.Ano), 1.0))


plt.savefig('sf_agric.png',bbox_inches='tight', transparent=True)
```
